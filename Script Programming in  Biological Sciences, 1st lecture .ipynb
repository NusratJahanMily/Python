```python
#while
# A loop which is repeated as long as the condition is met
# As soon as the condition is no longer true, exits the loop

count = 0
while count < 10: #as long as that count get 10,it will continue
    print (count)
    count += 2 # increasing by 2 number
print ("DONE\n") #\n means line break
```

```python
#not, and, or can be used inside conditional terms
#Order of precedence: not – and – or. To override, use parentheses

n = -2
item = 'number'
if (n > 10) and (item == 'number'): # only one fulfilled
    print ("rice fat!\n")
elif (n > 10) and (item != 'number'): #neither fulfilled
    print ("tea!\n")
else:
    print ("vegetable\n") #both condition fulfilled
```

```python
# Prints out every letter
string = 'rice and egg'     #A string is a sequence of characters #The loop goes one character at a time
for x in string:
    print(x)               #Each loop = one letter ; t → h → i → n → g → s
```

```python
#Loops can be nested, means loop inside loop
for x in range(5):
    for y in range (3):       # Outer loop (x) runs 10 times → 0 to 9  #For EACH x, inner loop (y) runs 5 times → 0 to 4
        print(x,y)           # For every x, you loop through ALL y ;x=0 → y=0,1,2,3,4 ; x=1 → y=0,1,2,3,4 
```

```python
for x in range(10):    #Skips 5, but loop continues
    if x == 5:
        continue
    print(x)
```

```python
for x in range(10):       #Stops at 5 → loop ends
    if x == 5:
        break
    print(x)
```

```python
sequence = 'gatacgatgcgttcagatgtcatgcagtcataatacgct'
pos = 0

while pos < (len(sequence) - 2):     #Because each codon needs 3 letters. So we stop before incomplete triplets
    end = pos + 3
    triplet = sequence[pos:end]

    if triplet == "atg":
        # check for start codon
        print("***")
        pos = end
        continue

    if (triplet == "tag") or (triplet == "taa") or (triplet == "tga"):
        # check for stop codon
        print("---\n")
        break

    pos = end
    print(triplet)
```

```python
#easier code for same result
sequence = 'gatacgatgcgttcagatgtcatgcagtcataatacgct'
stop_codons = {"tag", "taa", "tga"}

for i in range(0, len(sequence) - 2, 3):  #Automatically moves:  0 → 3 → 6 → 9 → ...
    codon = sequence[i:i+3]  #Takes 3 letters (a codon); Example: i = 0 → "gat" , i = 3 → "acg", i = 6 → "atg"

    if codon == "atg":
        print("***")

    elif codon in ["tag", "taa", "tga"]:
        print("---\n")
        break  #break → stop the loop immediately

    else:
        print(codon)
```
